### Procurando videos baseandose em palavra chave




```python
def bring_videos_base_on_key(youtube,keyword):

  request = youtube.search().list(
          part='snippet',
          maxResults=1,# default 5
          regionCode='BR',
          #The regionCode parameter instructs the API to return search results 
          #for videos that can be viewed in the specified country. 
          type='video',
          order='viewCount',
          q=keyword
      )
  response = request.execute()

  return(response)
```

### Procurando videos baseandose em video id

```python
def bring_videos_base_on_id(youtube,videoId):

  request = youtube.search().list(
          part='snippet',
          maxResults=2,# default 5
          regionCode='BR', # Vis√≠vel desde Brasil
          relatedToVideoId=videoId,
          #The regionCode parameter instructs the API to return search results
          #for videos that can be viewed in the specified country. 
          type='video',
          #order='rating'
      )
  response = request.execute()

  return(response)
```

Colocando agora isto numa fun√ß√£o, que simuaneamente converta o strint para lista, e a lusta para segundos

```python
def dura_to_segun(time):
  segundos=0
  import re 
  a=re.split('[f-u]+',time, flags=re.IGNORECASE)
  a=a[1:-1]
  a=[int(i) for i in a]
  if len(a)==2:
    segundos=a[0]*60+a[1]
  elif len(a)==3:
    segundos=a[2]+60*(a[1]+a[0]*60)
  else:
    segundos=a[0]
  
  return segundos

 

```
