Porem has horas espressadas por esta tabela estão no horario do fuso horário do meridiano de Greenwich, o qual está tres horas adiantado ao fusso horario de Brasilia.

Logo fazemos uma conversão para o horario Brasilia com a expectativa de que a maioria dos produtores de conteúdo trabalhem nesse horario.

Defimos uma função para fazer a soma aritmética da hora

```python
def to_brasilia_time(a):
  if a>=3:
    a=a-3
  else:
    a=24+a-3
  return(a)
```

```python
a=to_brasilia_time(-1)
a
```

```python
dados_datas_1['Hora']=dados_datas_1['Hora'].apply(to_brasilia_time)
dados_datas_1.head(1)
```
